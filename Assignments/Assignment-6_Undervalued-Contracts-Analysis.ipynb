# Assignment 6: Undervalued Contracts Analysis

## Objective
Use quantitative basketball metrics and salary data to identify three undervalued NBA player contracts and build a career, data-driven case for why each contract is a bargain.

## Instructions
- Identify 3 current NBA players whose contracts you believe are undervalued in the current season. For each player, you will
  - Report their current season salary
  - Provide 2-3 analytical bullets explaining why the contract is undervalued
  - Build a comparison table with that player and 2-3 comparable player seasons (from the last few years) that supports your argument.
- You must justify *undervalued* with data
