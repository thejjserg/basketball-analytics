## Evaluating Trades & Free Agency

### Guiding Question - What Do We Need?
- Scenario: You are a Top 10 Offense and Bottom 10 Defense
- You can add either
  - A high usage scorer with average defense
  - A lower-usage 3&D Wing with strong defense
- What information do you want before making this decision?

### Diagnosing Team Needs
- Start with the same framework used in evaluating opponents
  - OER, DER, NET
  - Four Factors
  - Shot Profile
  - Play Types
  - Star Players vs Role Players
- Question: What are we trying to accomplish with our roster with this move?

### Player-Level Impact for Their Role
- Core Concept: Evaluate players within the context of the role they will have on your team
- Example
  - Primary Creators - Playmaking, advantage creation weighted more heavily than rebounding
  - 3&D Wing - 3PT%, matchup difficulty, positional versatility weighted more than creating own shot

### Player-Level Immpact: Stylistic Fit & Scalability
- Fit
  - How a Player Likes to Play vs How Your Team Plays
  - Scheme fit on both ends (P/R Level, Help vs 1-on-1 Defense, Small vs Big Lineups)
- Scalability
  - Can they maintain or increase efficiency next to different stars?
  - Do their strengths complement your core players?
  - Can they contribute in the playoffs?

### Health, Reliability, and Playoff Availability
- Games played and minutes load over multiple seasons
- Injury history: Acute vs Chronic Issues and Recent Trends
- How have they handled the workloads you'll be asking of them?
- Playoff availability and performance history in high-leverage games
- *A great player who is rarely available is a risky investment*

## Salary Cap & Roster Efficiency

### Contract Lens: Short-Term vs Long-Term Deals
- Shorter Deals
  - More flexibility and easier to pivot if the fit is wrong
  - Less security for the player
  - Less control for the team
- Longer Deals
  - Lock-in value if the player outperforms their contract
  - Carries risk of decline, injury, or scheme changes
- Ideal: Match contract length to age, development curve, and team timeline

### Big Picture NBA Contract Considerations
- Age and expected development / decline inform what length and dollars you find a player to be worth
- Team Timeline: Rebuilding vs Contending Teams value contract types differently
- Option structures and guarantees
  - Team options and non-guarantees keep flexibility for teams
  - Player options give more control to the player
- Restricted vs Unrestricted Free Agency: value of matching rights and team control

### Salary Cap & Contract Types
- Core contract types
  - Rookie Scale - 1st Round Picks
  - Veteran Contracts
    - Maximum
    - Mid-Level Exception
    - Minimums
  - Two-Way
- Free Agency Status
  - Unrestricted
  - Restricted
- Don't visualize the Salary Cap in terms of $, rather, visualize it in terms of %

### Cap Concepts & Tools
- The NBA has a *soft* salary cap - teams can exceed the cap using exceptions
- **Luxury Tax**: impacts how many mid-level and max deals a team can realistically carry
- **Cap Holds**: Placeholder amounts that affect team salaries in the offseason
- Option years and guarantees
- **Aprons**: Limits to team building tools as you go further over the Salary Cap

## Undervalued Players & Market Inefficiencies

### Roster Efficiency
- Think of the roster as a portfolio of skills and salaries. The goal is to maximize wins within the constraints of the salary cap.
  - Star money for star-level impact
  - Mid-Tier deals for high-minute role players
  - Low cost depth and developmental bets to manage risk
- Redundancy for truly high value skills
- Complementary roster construction

### Identifying Undervalued Players
- **Goal**: Find players who provide more value than they cost
- Sources of Undervaluation:
  - Context
  - Development
  - Perception
  - Mis-evaluation
- Key Lenses:
  - Value of Winning
  - Opportunity
  - Development
  - Fit

### Players Who Impact Winning
- May appear in metrics beyond the box score, even with modest box score stats
- On-and-Off Data
- Real-Plus Minus Metrics
- Consistently in rotation in winning situations

### Players in Need of Opportunity
- Not given minutes due to depth, coach fit, scheme, etc.
- Role expansion when usage is capped behind stars
- Better scheme fit that unlocks strengths

### Value From Player Development Over Time
- Late bloomers and non-linear development
- Improved development system to build out their skillset
- More uncertainty with these players
- Players from other leagues - International and G-League Players whose skills translate to the Team System or League

### Data Patterns That Signal *Buy Low* Opportunities
- Historic success, down year candidates
  - Multiple seasons of strong efficiency or impact followed by one poor year
  - Evaluate whether the dip is injury-driven, context-driven, or simplae variance
- Process vs Outcome
  - Good shot quality and locations, but poor recent shooting results
  - Strong creation for others with low personal scoring
- Target players where the process is strong even if recent outcomes look poor
