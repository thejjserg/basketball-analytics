# Class 4

Net Rating is important

Possession Ends
1. Made Shot
2. Rebound
3. Turnover
4. Free Throws

Four Factors
1. Effective Field Goal % - Shootin Efficiency
2. Turnover Rater
3. Offensive Rebound Rate
4. Free Throw Rate




````python
import numpy as np
import pandas as pd

def projected_wins(
    team_efg, opp_efg,
    team_tov, opp_tov,
    team_orb, opp_orb,
    team_ft_rate, opp_ft_rate,
    games_played=82
):
    """
    Estimates projected wins using Dean Oliver's Four Factors model.

    Parameters
    ----------
    team_efg : float
        Team effective field goal percentage (0â€“1 scale)
    opp_efg : float
        Opponent effective field goal percentage (0â€“1 scale)
    team_tov : float
        Team turnover percentage (0â€“1 scale, e.g. 0.13 = 13%)
    opp_tov : float
        Opponent turnover percentage (0â€“1 scale)
    team_orb : float
        Team offensive rebound rate (0â€“1 scale)
    opp_orb : float
        Opponent offensive rebound rate (0â€“1 scale)
    team_ft_rate : float
        Team free throw rate (FTM / FGA)
    opp_ft_rate : float
        Opponent free throw rate (FTM / FGA)
    games_played : int
        Season length (default 82)

    Returns
    -------
    dict
        Dictionary containing:
        - Four Factors rating (composite)
        - Projected win percentage
        - Projected wins
    """

    # Compute net differences for each factor
    efg_diff = team_efg - opp_efg
    tov_diff = opp_tov - team_tov  # lower TOV is better
    orb_diff = team_orb - opp_orb
    ft_diff  = team_ft_rate - opp_ft_rate

    # Weighted sum (Dean Oliver's weighting)
    rating = (0.40 * efg_diff) + (0.25 * tov_diff) + (0.20 * orb_diff) + (0.15 * ft_diff)

    # Convert rating to win percentage (simple linear model approximation)
    win_pct = 0.5 + rating * 2.7  # scaling factor to match typical NBA ranges

    # Clamp between 0 and 1
    win_pct = max(0, min(1, win_pct))

    projected_wins = win_pct * games_played

    return {
        "Four Factors Rating": round(rating, 4),
        "Projected Win %": round(win_pct * 100, 2),
        "Projected Wins": round(projected_wins, 1)
    }
````

## Offensive and Defensive Efficiency
**Possession**
- Each time a team gets the ball starts a new possession
- Which team has the ball must change teams for a new possession to begin
- OREB = **Same Possession!**
- An Offensive Rebound is a **New Chance**, NOT a new possession
- Exact = Count!
- Estimated = FGA + 0.44 * FTA + TOV - OREB

**Pace**
- Pace is the number of possessions per 48 minutes played by a team
- Its purpose is to measure how many possessions on each side of the ball a team averages over a 48 minute period (length of a game)
- Formula: (48 * 0.5 * (Tm_Poss + Opp_Poss)) / Tm_MP

**Possessions & Efficiency**: Why Per 100?
- Normalize for pace differences
- Comparable rates unlocks apples-to-apples
- Rules of basketball - alternating possessions

**Offensive Efficiency (OER) and Offensive Rating (OffRtg)**
- Offensive Effieciency or Offensive Rating is the team's points scored per 100 possessions.
- Its purpose is to show how much a team scores, adjusting for pace. Measurement of how good an offense is
- Formula: 100 * (Points / Tm_Poss)

**Defensive Efficiency (DER) and Defensive Rating (DefRtg)**
- Defensive Efficiency or Defensive Rating is the opponent's points scored per 100 possessions
- Its purpose is to show how much a team allows, adjusting for pace. Measurement of how good a defense is
- Formula: 100 * (Opp_PTS / Opp_Poss)

**Net Rating**
- Net Rating is the point differential per 100 possessions
- Its purpose is to summarize team strength per-possession
- Formula: OER - DER

## Four Factors
**Ways a Possession Ends**
1. Made Shot
2. Rebound
3. Turnover
4. Free Throws

**Effective Field Goal % (eFG%)**
- eFG% is a player's shooting efficiency that adjusts for the fact that a made 3PT shot is wort 1.5x a made 2PT shot
- Its purpose is to improve upon FG% as a measure for shooting
- Formula: (FGM + 0.5 * 3PM) / FGA

**Offensive Rebound % (OREB%)**
- OREB% is a percentage of available offensive rebounds a team gets
- Its purpose is to normalize rebounding for number of shots and how well a team shoots
- Formula (Exact): OREB / OREB_Opps
- Formula (Estimated): OREB / (OREB + Opp_DREB)

**TOV%**
- TOV% is the turnovers per possession
- Its purpose is to normalize team turnovers for pace
- Formula: TOV / POSS

**FTA Rate**
- FTA Rate is the number of FTAs a team shoots per FGA
- Its purpose is to measure how often a team gets to the free throw line relative to how often they shoot
- Formula: FTA / FGA

**Four Factors - Dean Oliver's Weights**
1. eFG% - 40%
2. TOV% - 25%
3. OREB% - 20%
4. FTA Rate - 15%

## Analyzing In-Game Decisions
- What decisions can coaches make in-game?
  - Lineups
  - Matchups
  - Coverage Scheme
  - Offensive Play Calls
  - End-of-Game Tactics
- In-Game Decision Making
  - Situation
    - Score, Time, Possession, Fouls, Timeouts, Previous Plays
  - Options
    - Identify 2-3 realistic choices
  - Information
    - Priors (Season Data, Historical Performance)
    - New Evidence (What is Happening in this Game?)
  - Impact
    - Which option best improves outcomes (Points, Stops, Win Probability?)
- Balancing Priors vs New Information
  - **Priors**: What has been historically true
  - **New Information**: Tonight's Reality - Shooting Variance, Fouls, Injuries
  - Heuristics
    - Strong Prior + Noisy Game Data = Lean on Priors
    - Weak Prior + Clear Pattern Tonight = Respect the Game Data
    - Higher Leverage = Want Priors and Game Data Aligned
- Leverage of Moments
  - Not all decisions are equal in impact
  - Examples
    - Late Game Lineups
    - Major Coverage Changes
    - End of Game ATOs
    - Foul vs Defend, Quick 2 vs 3, etc.
  - **Goal**: Have simple, pre-built guidelines for these moments built on data.

## Tracking and Adjusting in Real Time
- Tracking Matchups in Real Time
  - Identify who guards
    - Star Player (Stylistic Match)
    - Primary Ballhandler
    - Bigs by Archetype
  - Track Outcomes by Matchup
  - **Objective**: Quickly find best defender and biggest mismatch
    - How does team scheme affect this?
    - How do we evaluate prior outcomes vs what we're seeing in-game?
- Measuring Coverage Performance
  - Pre-Game Priors
    - Opponent Efficiency vs Schemes
    - Prior Matchup Data
    - Opponent Tendencies vs Team Strengths
  - In-Game Data
    - Efficiency by Coverage
    - Shot Diet & Shot Quality Allowed
    - What can we find in the data that tells a story about what's happening in-game?
- Priors vs In-Game Data for Defense
  - Priors & Game Agree
    - If it's going well, then stay the course
    - If it's going poorly, then pivot *quickly*
  - Priors & Game Conflict
    - Is it a small sample?
    - How is team execution?
    - Process: Fix execution first, then adjust the scheme
- Triggers to Adjust in Real Time
  - Coverage expected efficiency well above pre-game expectations
  - Same action being used successfully against you repeatedly
  - Individual overmatched in size/speed/space
  - Weigh tradoffs: personnel fit, foul trouble, talent

## Optimizing Play Calling

### Pre-Scout: Offensive Game Plan
- Team Strengths
  - Best Actions
  - Best 1-on-1 Players vs Mismatches
- Opponent Tendencies
  - Man vs Zone (Type)
  - P/R Coverages
  - Switch Tendencies
  - Star Coverages
  - Help Tendencies
  - Elite vs Weak Defenders

### In-Game: What's Actually Working?
- Track Outcomes by Action
  - P/R Handler, Screener, and Pairs
  - Isolations & Post-Ups
  - The Different Defensive Coverages
  - Sample Size Considerations
- Key Questions to Consider
  - Are we getting our best actions?
  - Are we stuck running what the defense wants us to run?

### Priors vs In-Game For Play Calls
- Pre-Game Priors
  - Efficiency of Play Calls
  - Efficiency vs Coverages
  - Individuals vs Coverages
- In-Game Data
  - Play Call Outcomes
  - Expected Data vs Actual Efficiency
- Final Decision
  - Continue with our **baseline strategy** or **change course** with what's happening tonight?

## Lineup Optimization and Rotations
- Pre-Game Lineup Priors
- Use Priors to **build** baseline rotations and planned closing lineups
  - For Each Lineup Bucket, calculate
    - OER, DER, NET Rating
    - Minutes Played (Sample)
    - Four Factors & Shot Diet - What is this lineup generating and taking away?
  - Go-To Lineups by Strengths
    - Shooting (Our Best 3s, Best FT%)
    - Big Lineup
    - Small Lineup
    - Defensive Focused
- **Lineup Priors Rule of Thumb**: *the best lineups include all of your best players!*
- In-Game Lineup Signals
  - Player Context: Foul Trouble, Fatigue
  - Stint-Level Performance
    - Score Margin in Each Stint
    - Shot Quality Generated and Allowed (Expected Efficiency)
    - Four Factors
  - Opponent Lineup Considerations
