# Class 1 - Introduction to Basketball Analytics

## Key Applications
- Game Strategy
1. Pre-Game: Understanding opponent strtegnths, weaknesses, and tendencies. Not a substitute for film though
2. In-Game: Influencing strategy changes with data
3. Post Game: Analyzing strategy and execution after the game

- Team Evaluation
1. Individual roles: How do we optimize the players we have?
2. Lineups: What combinations of players do we want to play? Which do we want to avoid?
3. Post-Game: Why did we win or lose the game? Did we execute our gameplan?

- Scouting
1. Talent & Skill: How good is a player? What are their strengths and weaknesses?
2. Player Value: How much is a player worth? How do they fit within our salary cap/budget?
3. Fit: How does a player fit on court with our roster?

- Player Development
1. Skill Identifiaction: In what areas does this player need to improve?
2. Prioritization: What should a player be working on?
3. Measuring Progress: How can we best measure a player's improvement over team?

Notes:
- Importance is determined by where the team is and where we are in the calendar.
- Measuring Players for players that don't get a lot of time. If it's important, then how do you measure it?
